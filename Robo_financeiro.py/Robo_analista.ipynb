```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import mplcyberpunk
```

```python
tickes = ["XRP-USD","SOL-USD"]

dados = yf.download(tickes, period= "6mo", auto_adjust= False)
dados
```

```python
dados_de_fechamento = dados["Adj Close"]
dados_de_fechamento
```

```python
dados_de_fechamento.columns = ["Solana","XRP"]
dados_de_fechamento
```

```python
plt.style.use("cyberpunk")
```

```python
plt.plot(dados_de_fechamento["Solana"])
plt.title("Solana")

```

```python
plt.plot(dados_de_fechamento["XRP"])
plt.title("XRP")
```

```python
retorno_diario = dados_de_fechamento.pct_change()
retorno_diario
```

```python
retorno_Solana = retorno_diario["Solana"].iloc[-1]
retorno_XRP = retorno_diario["XRP"].iloc[-1]
```

```python
retorno_Sol = str(round(retorno_Solana * 100, 2)) + "%"
retorno_XRP = str(round(retorno_XRP * 100, 2)) + "%"
```

```python
print(retorno_Sol)
print(retorno_XRP)
```
